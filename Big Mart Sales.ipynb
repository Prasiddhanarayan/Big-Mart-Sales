```python
import pandas as pd
import numpy as np

#Read files:
train = pd.read_csv("C:/Users/hp/Desktop/Big_Mart/train.csv")
test = pd.read_csv("C:/Users/hp/Desktop/Big_Mart/test.csv")
```

```python
data = pd.concat([train, test],ignore_index=True)
print (train.shape, test.shape, data.shape)
```

```python
data.dtypes
```

```python
categories = [x for x in data.dtypes.index if data.dtypes[x]=='object']
#Exclude ID cols and source:
categories = [x for x in categories if x not in ['Item_Identifier','Outlet_Identifier','source']]
#Print frequency of categories
for col in categories:
    print ('\nFrequency of Categories for varible')
    print(col)
    print (data[col].value_counts())
```

```python
data['Item_Identifier'].value_counts()
```

```python
data['Outlet_Size'].fillna('Medium', inplace=True)
```

```python
data['Item_Weight'].fillna(data['Item_Weight'].mean(), inplace=True)
```

```python
data.pivot_table(values='Item_Outlet_Sales',index='Outlet_Type')
```

```python
data['ItemVisibility']= data['Item_Visibility'].map(lambda d: data['Item_Visibility'].mean() if d==0 else data['Item_Visibility'])
```

```python
data.drop(['ItemVisibility'],axis=1, inplace=True)
```

```python
data.head()
```

```python
data['Item_Visibility'].replace(0.00000,np.nan, inplace=True)
```

```python
data.isnull().sum()
```

```python
a=data['Item_Visibility'].mean()
```

```python
data['Item_Visibility'].fillna(a,inplace=True)
```

```python
data.head()
```

```python
data['Item_Fat_Content'].replace('LF','Low Fat', inplace=True)
data['Item_Fat_Content'].replace('low fat','Low Fat', inplace=True)
data['Item_Fat_Content'].replace('reg','Regular', inplace=True)
```

```python
data['Outlet_Years']=2018-data['Outlet_Establishment_Year']
data.drop('Outlet_Establishment_Year', axis=1, inplace=True)
```

```python
data['Item_Type_New'] = data['Item_Identifier'].apply(lambda x: x[0:2])
```

```python
data['Item_Type_New'] = data['Item_Type_New'].map({'FD': 'Food',
                                              'DR': 'Drinks',
                                              'NC': 'Non-Consumable'})
```

```python
data.drop('Item_Identifier', axis=1, inplace=True)
```

```python
data.loc[data['Item_Type_New']=="Non-Consumable",'Item_Fat_Content'] = "Non-Edible"
```

```python
data.dtypes
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Outlet_Type', 'Item_Type_New','Outlet_Identifier'])
```

```python
data.head()
```

```python
data.dtypes
```

```python
data = pd.get_dummies(data, columns=['Item_Type'])
```

```python
data.head()
```

```python
Y=train['Item_Outlet_Sales']
```

```python
data.drop('Item_Outlet_Sales', axis=1, inplace=True)
```

```python
X = data.head(8523)
X_test = data.iloc[8523:]
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression(normalize=True)
```

```python
lr.fit(X, Y)
```

```python
lr_pred = lr.predict(X_test)
```

```python
lr_accuracy = round(lr.score(X,Y) * 100,2)
```

```python
lr_accuracy
```

```python
#Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor(max_depth=15,min_samples_leaf=100)
```

```python
tree.fit(X,Y)
```

```python
tree_pred = tree.predict(X_test)
```

```python
tree_accuracy = round(tree.score(X,Y),2)
tree_accuracy
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=400,max_depth=6, min_samples_leaf=100,n_jobs=4)
```

```python
rf.fit(X,Y)
```

```python
rf_pred = rf.predict(X_test)
```

```python
rf_accuracy = round(rf.score(X,Y),2)
rf_accuracy
```
